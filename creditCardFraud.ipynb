##Importing Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score
```

##Data Collection

```python
card_data = pd.read_csv("/content/creditcard[1].csv")
```

##Data Pre-Processing

```python
card_data.head()
```

```python
card_data.tail()
```

```python
card_data.shape
```

```python
card_data.info()
```

##Evaluation of Missing Values

```python
card_data.isna().sum()
```

```python
card_data.dropna(inplace = True)
```

##Data Analysis

```python
card_data.describe()
```

```python
card_data['Class'].value_counts()
```

```python
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

##Splitting The data into Train and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

##Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

##Accuracy Score Evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
